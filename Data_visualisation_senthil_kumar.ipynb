> ## Bivariate Analysis

1. the first value must be categorical data
2. hue must be categorical data

```python
import scipy.stats as stats

sb.jointplot(traffic15['Happiness Score'], traffic15['GDP per Capita'])


# scatter plot uses continuous value
```

```python
import scipy.stats as stats

j = sb.jointplot(traffic15['Happiness Score'], traffic15['GDP per Capita'])
j.annotate(stats.pearsonr)

# we uses pearsonr corelation ( we have two values to corelate)
# scatter plot uses continuous value
```

```python
# regression plot shows the optimal line from the data 
```

```python
sb.regplot(traffic15['Happiness Score'], traffic15['GDP per Capita'])
```

```python
# It perform all the plots from all the data


sb.pairplot(traffic15)
```

```python
# boxplot in bivariate


plt.figure(figsize=(15,8))       # change the size of the figure
sb.boxplot(x = traffic15['Region'], y = traffic15['GDP per Capita'])
plt.xticks(rotation=90)       # change the names of the x axis in horizontal
#plt.yticks(rotation=90)    #   change the names of the y axis in horizontal



# if bar height is more than that means there is more deviation.
# so the north america having the good GDP, becoz they earn 
```

```python
plt.figure(figsize=(15,8))       # change the size of the figure

sb.violinplot(x = traffic15['Region'], y = traffic15['GDP per Capita'])
plt.xticks(rotation=90)      


# in southern asia, there is more poor and same no. of rich so there is more stretching, and people are ranging as per the income 
# in north america the plot width is more that means there is a lots of people in that range
```
